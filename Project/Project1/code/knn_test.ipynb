```python
import numpy as np # linear algebra
import pandas as pd # data processing
from fancyimpute import  KNN
import math
```

```python
#Reading naidx file

file_name = './AP1/test_data/test_data/naidx.csv'
df_naidx=pd.read_csv(file_name)
len_naidx=len(df_naidx)
```

```python
#Reading files for imputation

file_name = './AP1/test_data/test_data/test_with_missing/{}.csv'
df_list = []

for i in range(1, 2268):
    df=pd.read_csv(file_name.format(i))
    
    #Applying KNN per patient using k=5
    train_cols = list(df)
    df_new=pd.DataFrame(KNN(k=5).fit_transform(df))
    df_new.columns = train_cols
    
    #Writing to files
    filename='./AP1/test_data/test_data/test_imputed/{}.csv'
    df_new.to_csv(filename.format(i), sep='\t')
```
